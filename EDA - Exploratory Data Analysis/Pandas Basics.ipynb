## 13. Merge

```python
# Series 1: Country and GDP (in billions USD)
gdp_data = {
    'United States': 21430,
    'China': 14340,
    'Japan': 5080,
    'Germany': 3860,
    'India': 2870
}
gdp_series = pd.Series(gdp_data)

# Series 2: Country and Population (in millions)
population_data = {
    'China': 1410,
    'India': 1390,
    'United States': 331,
    'Indonesia': 273,
    'Pakistan': 220
}
population_series = pd.Series(population_data)
    
```

```python
df1 = pd.DataFrame(gdp_series).reset_index()
```

```python
df1.columns = ['Country','GDP_Per_Capita']
```

```python
df2 = pd.DataFrame(population_series).reset_index()
```

```python
df2.columns = ['Country','Population']
```

```python
pd.merge(df1, df2, on = 'Country' , how = 'inner')
```

```python
pd.merge(df1, df2, on = 'Country' , how = 'left')
```

```python
pd.merge(df1, df2, on = 'Country' , how = 'outer')
```

```python
pd.merge(df1, df2, on = 'Country' , how = 'right')
```
