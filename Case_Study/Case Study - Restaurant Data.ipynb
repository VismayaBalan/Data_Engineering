## 6. Approx of Restaurant Cost (2 People)

```python
zomato_df['approx_cost(for two people)'].unique()
```

```python

zomato_df['approx_cost(for two people)'] = zomato_df['approx_cost(for two people)'].str.replace(',', '').astype('float64')

```

## Check if it is skewed ?

```python
import warnings
warnings.filterwarnings('ignore')
plt.figure(figsize = (14,8))
sns.set_style('whitegrid')
sns.distplot( zomato_df['approx_cost(for two people)'])
```

*  Yes, the approx cost for 2 person is skewed towards positive. Hence positive skewness

```python
plt.figure(figsize = (14,8))
sns.set_style('whitegrid')
sns.boxplot( zomato_df['approx_cost(for two people)'],orient = 'h',color = 'Red')
```

```python
zomato_df['approx_cost(for two people)'].median()
```

```python
Q1 = zomato_df['approx_cost(for two people)'].quantile(0.25)
Q2 = zomato_df['approx_cost(for two people)'].quantile(0.50)
Q3 = zomato_df['approx_cost(for two people)'].quantile(0.75)
print("First Quartile: " , Q1, "\nMedian: ", Q2, "\nThird Quartile: ", Q3)
```

```python
IQR = Q3 - Q1
print("Inter quartile range of cost for 2 people: ", IQR)
```

* Approx cost of two people range from 300 to 650 INR with median cost approx to Rs. 400. Yes, the approx cost for 2 person is skewed towards positive. Hence positive skewness
